```python
import pandas as pd
data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

data2 = {
  "name": ["Sally", "Peter", "Micky"],
  "age": [77, 44, 22]
}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
newdf=df1.merge(df2,how='right')
```

```python
newdf
```

```python
df1 = pd.DataFrame({'name': ['Alice']})
df2 = pd.DataFrame({'name': ['Bob']})

result = pd.concat([df1, df2])
result
```

```python
#1. Simple Concat (Rows)
#Task: Combine them into a single DataFrame with all 4 names.
df1=pd.DataFrame({'name':['Alice','Bob']})
df2=pd.DataFrame({'name':['Charlie','David']})
```

```python
result=pd.concat([df1,df2],ignore_index=True)
result
#alternative: result = result.reset_index(drop=True)
```

```python
#2 Simple Concat (Columns)
df1 = pd.DataFrame({'name': ['Alice', 'Bob']})
df2 = pd.DataFrame({'age': [25, 30]})
result=pd.concat([df1,df2],axis=1)
result
```

```python
#3 Basic Merge
df1 = pd.DataFrame({'id': [1, 2], 'name': ['Alice', 'Bob']})
df2 = pd.DataFrame({'id': [1, 2], 'score': [90, 85]})
result=df1.merge(df2,how='inner',on='id')
result
#alternative: result = df1.merge(df2, on='id') since inner is the default in pandas
```

```python
#4 Left Join Practice
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'id': [1, 2], 'score': [90, 85]})
result=df1.merge(df2,how='left',on='id')
result
```

```python
#5 Outer Join
result=df1.merge(df2,how='outer',on='id')
result
```

```python
#6 Concat with Different Columns
df1 = pd.DataFrame({'name': ['Alice'], 'age': [25]})
df2 = pd.DataFrame({'name': ['Bob'], 'score': [90]})
result=pd.concat([df1,df2], ignore_index=True)
result
```

```python
#7 Multi-Key Merge
df1 = pd.DataFrame({
    'id': [1, 1, 2],
    'subject': ['Math', 'English', 'Math'],
    'score': [90, 85, 88]
})

df2 = pd.DataFrame({
    'id': [1, 2],
    'name': ['Alice', 'Bob']
})
result=df1.merge(df2,on='id')
result
```

```python
#8 Concat with Index Reset
df1 = pd.DataFrame({'name': ['Alice', 'Bob']}, index=[0, 1])
df2 = pd.DataFrame({'name': ['Charlie']}, index=[0])
result=pd.concat([df1,df2],ignore_index=True) #is this resetting index?
result
```

```python
#9 Real-World Scenario (Sales Data)
jan = pd.DataFrame({'product': ['A', 'B'], 'sales': [100, 150]})
feb = pd.DataFrame({'product': ['A', 'B'], 'sales': [120, 130]})
combined_months=pd.concat([jan,feb],ignore_index=True)
combined_months['month'] = ['Jan'] * len(jan) + ['Feb'] * len(feb)
combined_months
```

```python
#10 Merge + Concat Combined
users = pd.DataFrame({'id': [1, 2], 'name': ['Alice', 'Bob']})
orders1 = pd.DataFrame({'id': [1], 'order': ['Book']})
orders2 = pd.DataFrame({'id': [2], 'order': ['Pen']})
orders_all=pd.concat([orders1,orders2],ignore_index=True)
orders_all
```

```python
merged=orders_all.merge(users,how='left',on='id')
merged
```

```python
#11 Detect Differences After Concat
#given two datasets old vs new
old = pd.DataFrame({'id': [1, 2], 'value': [10, 20]})
new = pd.DataFrame({'id': [2, 3], 'value': [25, 30]})
combined = old.merge(new, on='id', how='outer', indicator=True)
combined
```

```python
#now to extract the desired rows:
#New rows
combined[combined['_merge'] == 'right_only']
```

```python
#Updated rows
both = combined[combined['_merge'] == 'both']
both[both['value_x'] != both['value_y']]
```
