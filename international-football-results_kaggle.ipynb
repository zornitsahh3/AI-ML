```python
df.groupby('year')['home_score'].mean()
```

```python
home = df[['date', 'home_team', 'home_score']].rename(
    columns={'home_team': 'team', 'home_score': 'score'}
)

away = df[['date', 'away_team', 'away_score']].rename(
    columns={'away_team': 'team', 'away_score': 'score'}
)

long_df = pd.concat([home, away])
#this way we have one row per team per match
```

```python
#Some questions:
#1: Does playing at home increase win probability?
df['home_win'] = (df['home_score'] > df['away_score']).astype(int)
df['home_win'].mean() #hence the probability that home team wins is 49%
```

```python
#2: Has scoring increased over time?
df['goal_diff']=df['home_score']-df['away_score']
df['home_win']=(df['goal_diff']>0).astype(int)
df['year'] = pd.to_datetime(df['date']).dt.year
df['total_goals'] = df['home_score'] + df['away_score']
df.groupby('year')['total_goals'].mean().max()
```

```python
#3: What’s the distribution of goal differences?
df['goal_diff'] = df['home_score'] - df['away_score']
df['goal_diff'].describe()
df['goal_diff'].value_counts().sort_index() #the goal difference, the number the specified goal appears
```

```python
df['goal_diff'].hist() #Does this follow a known distribution?
#Answer:The empirical distribution is centered around zero with heavier mass on small differences. It resembles a discrete, slightly skewed distribution rather than a perfect normal distribution
```

```python
import seaborn as sns
sns.histplot(df['goal_diff'], kde=True)
#Big spike at 0. The tallest bar is at goal difference 0: many games end in a draw.
#Close games (difference ±1, ±2) are common; blowouts (large positive or negative differences) are rare.
```
